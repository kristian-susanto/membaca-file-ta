```python
from zipfile import ZipFile
import pandas as pd
import os

# 1. Deklarasi nama file ZIP
zip_file_name = "Mingguan.zip"

# 2. Ekstrak file ZIP ke folder 'csv_folder'
with ZipFile(zip_file_name, 'r') as zip_ref:
    zip_ref.extractall("csv_folder")

# 3. Baca dan tampilkan file di dalamnya
folder_path = "csv_folder"
for i in range(1, 22):
    file_path = os.path.join(folder_path, f"{i}.csv")
    if os.path.exists(file_path):
        print(f"\n--- Isi dari {i}.csv ---")
        df = pd.read_csv(file_path)
        display(df)
    else:
        print(f"File {i}.csv tidak ditemukan.")
```
